```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def bar_plot_counts(df, col):
  fig = plt.figure(figsize = (8, 3))
  unique_values = df[col].unique().tolist()
  counts = [np.sum(df[col] == u) for u in unique_values]
  plt.bar(list(range(len(counts))), counts)
  plt.xticks(list(range(len(counts))), labels = unique_values)
  plt.ylabel("count")
  plt.title(f"Distribution of values for {col} column")
  plt.show()

def hist_plot(df, col):
  fig = plt.figure(figsize = (8, 3))
  plt.hist(df[col].values, bins = 100)
  plt.ylabel("count")
  plt.title(f"Distribution of values for {col} column")
  plt.show()

def plot_nan_dist(df):
  nan_avg = df.isna().mean().values
  fig = plt.figure(figsize = (8, 3))
  plt.hist(nan_avg, bins = 20)
  plt.ylabel("count")
  plt.xlabel("Fraction of nan values")
  plt.title("Distribution of nan values in columns")
  plt.show()

def plot_dtypes(df):
  fig = plt.figure(figsize = (8, 3))
  dtypes = df.dtypes
  unique_values = dtypes.unique().tolist()
  counts = [np.sum(dtypes == u) for u in unique_values]
  plt.bar(list(range(len(counts))), counts)
  plt.xticks(list(range(len(counts))), labels = unique_values)
  plt.ylabel("count")
  plt.title(f"Distribution dtypes of columns")
  plt.show()

def plot_unique_counts(df):
  fig = plt.figure(figsize = (8, 3))
  counts = []
  for col in df.columns:
    if df[col].dtype == "object":
      counts.append(len(df[col].unique()))
  plt.hist(counts, bins = max(counts))
  plt.ylabel("count")
  plt.xlabel("Number of unique values")
  plt.title("Histogram of unique values per column")
  plt.show()
```

```python
path = "/content/drive/MyDrive/courses/ai_ethics/school_eval/data/microdados_saeb_2019/DADOS/TS_ESCOLA.csv"
df_escola = pd.read_csv(path, encoding = "latin1", sep = ";")
print(f"Total rows: {df_escola.shape[0]}")
print(f"Total columns: {df_escola.shape[1]}")
print("Columns names:", df_escola.columns.tolist())
print(f"Unique schools: {len(df_escola.ID_ESCOLA.unique())}")
print(f"Unique cities: {len(df_escola.ID_MUNICIPIO.unique())}")
print(f"Collected years: {df_escola.ID_SAEB.unique().tolist()}")
```

```python
path = "/content/drive/MyDrive/courses/ai_ethics/school_eval/data/microdados_saeb_2019/DADOS/TS_DIRETOR.csv"
df_diretor = pd.read_csv(path, encoding = "latin1", sep = ";")
print(f"Total rows: {df_diretor.shape[0]}")
print(f"Total columns: {df_diretor.shape[1]}")
print("Columns names:", df_diretor.columns.tolist())
print(f"Unique schools: {len(df_diretor.ID_ESCOLA.unique())}")
print(f"Unique cities: {len(df_diretor.ID_MUNICIPIO.unique())}")
print(f"Collected years: {df_diretor.ID_SAEB.unique().tolist()}")
```

```python
# some directors did not answer the formulary, will remove than from the dataframe
df_diretor = df_diretor[df_diretor.IN_PREENCHIMENTO_QUESTIONARIO == 1]

# I can join the tables "escola" and "diretor", fill schools have more than one director
# I will just pick the first row from each of these schools

# getting only one director by school
df_diretor = df_diretor.groupby("ID_ESCOLA").agg("first").reset_index()

# now join dataframes
df = pd.merge(
    df_escola,
    df_diretor,
    on = "ID_ESCOLA",
    how = "outer",
    suffixes = ("", "_copy")
)
df = df.drop(columns = [col for col in df.columns if "_copy" in col])

# we will focus on final years of basic education
# so we remove columns related to the saeb of other yeras
columns_to_remove = [col for col in df.columns if "Nivel_" in col] + [
  "MEDIA_5EF_LP",
  "MEDIA_5EF_MT",
  "MEDIA_EMT_LP",
  "MEDIA_EMT_MT",
  "MEDIA_EMI_LP",
  "MEDIA_EMI_MT",
  "MEDIA_EM_LP",
  "MEDIA_EM_MT",
  "PC_FORMACAO_DOCENTE_INICIAL",
  "PC_FORMACAO_DOCENTE_MEDIO",
  "NU_MATRICULADOS_CENSO_5EF",
  "NU_PRESENTES_5EF",
  "TAXA_PARTICIPACAO_5EF",
  "NU_MATRICULADOS_CENSO_EMT",
  "NU_PRESENTES_EMT",
  "TAXA_PARTICIPACAO_EMT",
  "NU_MATRICULADOS_CENSO_EMI",
  "NU_PRESENTES_EMI",
  "TAXA_PARTICIPACAO_EMI",
  "NU_MATRICULADOS_CENSO_EM",
  "NU_PRESENTES_EM",
  "TAXA_PARTICIPACAO_EM",
]
df = df.drop(columns = columns_to_remove)


# many schools does not have the final years of basic education, so we will filter the ones that have
df = df[df.TX_RESP_Q037 == "A"]
```

```python
print(f"Total rows: {df.shape[0]}")
print(f"Total columns: {df.shape[1]}")
print("Columns names:", str(df.columns.tolist()))
print(f"Unique schools: {len(df.ID_ESCOLA.unique())}")
print(f"Unique cities: {len(df.ID_MUNICIPIO.unique())}")
```

```python
plot_nan_dist(df)
plot_dtypes(df)
plot_unique_counts(df)
```

```python
hist_plot(df, "MEDIA_9EF_LP")
hist_plot(df, "MEDIA_9EF_MT")
```
